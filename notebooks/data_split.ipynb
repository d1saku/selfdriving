```python
import pandas as pd

df = pd.read_csv('../data/training_norm.csv')
df
```

```python
from sklearn.model_selection import train_test_split

train_df, val_df = train_test_split(
    df,
    test_size=0.2,
    random_state=0,
    shuffle=True
)

train_df.to_csv('../data/train.csv')
val_df.to_csv('../data/validation.csv')
```
